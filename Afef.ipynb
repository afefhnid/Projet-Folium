# Import Json 

```python
import requests
import json
import csv 
import folium

url = 'https://lefooding.com/fr/search?type=restaurant&filters%5Bgeo%5D%5Bcity%5D%5B%5D=526ba3f07f55b7f39a007e28&page=1&sort=&per_page=460&total_pages=94'
r= requests.get(url)
print(url,r.status_code)
```

```python
r.json()
```

# Convert json in to txt file

```python
data =r.json()
with open('data.txt', 'w') as f:
    json.dump(data, f)
```

# Collect Json

```python
data['results'][:1]
df=data['results']
```

```python
len(data['results'])
```

```python
# nom addresse phone_num cravings price 
l_name=[]
l_phone=[]
l_cravings=[]
l_specialitees=[]
l_price=[]
l_position=[]
l_postal_code=[]
l_city=[]
l_street_name=[]
l_street_name_1=[]
l_log=[]
l_lat=[]
for elem in data['results']:
    l_name.append(elem['name'])
    l_log.append(elem['address']['position'][0])
    l_lat.append(elem['address']['position'][1])
    l_postal_code.append(elem['address']['postal_code'])
    l_street_name_1.append(elem['address']['street_name_1'])
    l_city.append(elem['address']['city'])
    l_phone.append(elem['phone_num'])
    l_cravings.append(','.join(elem['cravings']))
    l_specialitees.append(elem['cravings'])
    l_price.append(elem['price'])
    
    
print (len(l_name),len(l_phone),len(l_cravings),len(l_price))
```

```python
#Création d'un DataFrame

import pandas as pd

df_resto=pd.DataFrame({'Nom':l_name,'Téléphone':l_phone,'Spécialitées':l_cravings,'Prix':l_price,'City':l_city,'postal_code':l_postal_code,'Street_name_1':l_street_name_1,'Latitude':l_log,'Longitude':l_lat})

#Affichage de  lignes au hasard
df_resto.sample(10)
```
